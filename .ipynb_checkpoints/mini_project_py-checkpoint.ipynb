# <center>DATA CLEANING AREA</center>

```python
raw_data.columns = [raw_data.columns[i].lower().replace(' ', '_') for i in range(len(raw_data.columns))]
```

```python
data=raw_data.dropna()
```

### <center>DEFINE FUNCTIONS AREA</center>

```python
#This function converts the height in feet and inches to meters
def ft_inch_to_meter(i):
    i=i.replace('"','').split("'")
    feet=float(i[0])
    inch=float(i[1])
    meter = ((feet*0.3048) + (inch*0.0254))
    meter = round(meter,2)
    return meter
```

```python
#This function converts the weight in lbs to kgs
def lbs_to_kg(i):
    i = i.replace('lbs','')
    i = int(i)
    kg = round(i*0.45359237,2)
    return kg
```

```python
#This function calculates the BMI (body mass index)
def bmi(w,h):
    return round(w/(h*h),2)
```

```python
#This function turns the wage values to numbers
def wage_to_number(i):
    i=i.replace('€','').replace('K','000')
    i=float(i)
    return i
```

### <center>CONVERTION AREA</center>

```python
# Here we're converting the values of weight to kgs in data
for i in range(len(data['weight'])):
    data['weight'].iloc[i]=lbs_to_kg(data['weight'].iloc[i])
```

```python
# Here we're converting the values of height to meters in data
for i in range(len(data['height'])):
    data['height'].iloc[i]=ft_inch_to_meter(data['height'].iloc[i])
```

```python
# Renaming the column names accordingly to the changes made 
data=data.rename(columns = {'height':'height_(m)'})
data=data.rename(columns = {'weight':'weight_(kg)'})
```

```python
# Here we're converting the wage values to numbers
for i in range(len(data['wage'])):
    data['wage'].iloc[i]=wage_to_number(data['wage'].iloc[i])
```

### <center>CALCULATION AREA</center>

```python
data['bmi']=''
for i in range(len(data)):
    data['bmi'].iloc[i] =bmi(data['weight_(kg)'].iloc[i],data['height_(m)'].iloc[i])
```
